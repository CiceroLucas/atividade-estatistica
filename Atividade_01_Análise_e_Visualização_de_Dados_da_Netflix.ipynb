## Categorização da Netflix

O sistema de categorização da Netflix permite agrupar mais de uma categoria para o mesmo filme, logo utilizaremos de duas formas para fazermos a análise.
A primeira, vamos comparar com os dados aninhados, conforme a categorização netflix.

```python
df['Type'].value_counts().head(5).plot.barh(title='Categoria x Qtd. Produções')
```

Para aquelas que tem duas categorias ou mais, apenas a primeira categoria, supondo ser a mais relevante.

```python
first_data_type = df['Type'].str.split(',', expand=True)[0]
first_data_type.value_counts().head(5).plot.area(title='Categoria x Qtd. Produções', rot=10)
```

## Cálculo das estatísticas

```python
movies_df = df[df['Category'] == 'Movie'].copy()

movies_df['Duration'] = movies_df['Duration'].str.replace(' min', '').astype(float)

# estatísticas descritivas da duração dos filmes
stats = movies_df['Duration'].describe()

# variância e desvio padrão
variance = movies_df['Duration'].var()
std_dev = movies_df['Duration'].std()

stats, variance, std_dev
```
